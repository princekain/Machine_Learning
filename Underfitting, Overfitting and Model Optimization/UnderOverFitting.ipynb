```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize 
import numpy as np
import pandas as pd
df_train=pd.read_csv("regression_train.csv")
df_val=pd.read_csv("regression_val.csv")
train_error_list=np.zeros(8)
val_error_list=np.zeros(8)

x_train_1 = df_train['x'].values
y_train_1 = df_train['y'].values
x_val_1 = df_val['x'].values
y_val_1 = df_val['y'].values



plt.scatter(x_train_1,y_train_1, color='purple', alpha=0.5)
plt.scatter(x_val_1,y_val_1, color='blue',alpha=0.5)
plt.show()
```

```python
L=list(np.arange(0,8,1))
```

1. FOR TRAINING

```python
def mse_loss(params, x_train_1, y_train_1,i1): 
    beta=np.zeros(i1+2)
    y_hat=np.zeros(len(x_train_1))
    alpha=list(np.arange(0,i1+2,1))
    beta = params 
    for j in alpha:
        y_hat = y_hat+beta[j]*pow(x_train_1,j)
    mse=(y_train_1-y_hat)*(y_train_1-y_hat)
    return np.mean(mse)
```

```python
for i in L:
    alpha=np.zeros(i + 2)
    initial_guess = alpha
    result_mse = minimize(mse_loss, initial_guess, args=(x_train_1, y_train_1,i)) 
    train_error_list[i] = result_mse.fun
    val_error_list[i]=mse_loss(result_mse.x,x_val_1,y_val_1,i)
    #print(result_mse.x,"\n")
    plt.figure(figsize=(8, 5))
    
    # Plot actual data points
    plt.scatter(x_train_1, y_train_1, color='purple', alpha=0.4, label='Train Data')
    plt.scatter(x_val_1, y_val_1, color='blue', alpha=0.4, label='Val Data')
    
    # Generate smooth X values for the curve
    x_range = np.linspace(min(x_train_1.min(), x_val_1.min()), 
                         max(x_train_1.max(), x_val_1.max()), 100)
    
    # Calculate predicted Y for the curve using the optimized parameters
    y_plot = np.zeros(len(x_range))
    beta_opt = result_mse.x
    for j in range(i + 2):
        y_plot += beta_opt[j] * (x_range**j)
        
    # Plot the fitted line
    plt.plot(x_range, y_plot, color='red', linewidth=2, label=f'Degree {i+1} Fit')
    
    plt.title(f"Polynomial Regression (Degree {i+1})\nTrain MSE: {train_error_list[i]:.4f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
```

1. FOR VALIDATION

```python
import matplotlib.pyplot as plt

# Degrees of x from 1 to 8
degrees = list(range(1, 9))

# Plotting the training error
plt.plot(degrees, train_error_list, label='Training Error', marker='o', color='blue')

# Plotting the validation error
plt.plot(degrees, val_error_list, label='Validation Error', marker='x', color='red')

# Adding labels and title
plt.xlabel('Degree of Polynomial (x)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Training vs Validation Error')
plt.xticks(degrees) # Ensures all degree values (1-8) are shown on x-axis
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Save the plot
plt.show()
```
