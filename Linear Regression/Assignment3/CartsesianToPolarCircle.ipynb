```python
from scipy.optimize import minimize 
import numpy as np
import pandas as pd
df=pd.read_csv("dataset A.csv")
x = df['X'].values
y = df['Y'].values
r=pow((x*x+y*y+),0.5)
```

```python
def mse_lossSquare(params,x,y,r): 
    R= params 
    noise = r - R
    mseSqr= noise**2
    return np.mean(mseSqr)

```

```python
initial_guess = [0.01] 
result_mse = minimize(mse_lossSquare, initial_guess, args=(x,y,r)) 
R = result_mse.x 
print(R)


```

```python
import matplotlib.pyplot as plt
y1 = pow((R**2-(df['X'])**2),0.5) 
y2 =-y1
plt.figure(figsize=(10,6)) 
plt.scatter(df['X'], df['Y'], color='gray', alpha=0.5, label='Data') 
plt.scatter(df['X'], y1, color='red', label='') 
plt.scatter(df['X'], y2, color='green', label='', linestyle='--') 
plt.legend() 
plt.title("Polynomial Regression Comparison") 
plt.show() 
```
